## Problema

Una fábrica produce dos tipos de recipientes, K y L, para lo cual utiliza dos máquinas, $M_1$ y $M_2$. Para producir un recipiente K, $M_1$ necesita 2 minutos y $M_2$, 4 minutos. De manera semejante, L ocupa 8 minutos a $M_1$ y 4 minutos a $M_2$. La ganancia neta para un recipiente K es de 29 pesos y para uno L es de 45 pesos. De manera formal:

#### Variables de decisión:

Número de recipientes K, $x_1$ y L, $x_2$.

#### Función Objetivo:

No negatividad: $x_1 \geq 0$, $x_2 \geq 0$

#### Ejercicio 1

Una compañía fabrica dos tipos de motores pequeños con combustible sólido. Con el motor A gana $ \$  3$ por motor y con el motor B gana $\$ 4$ por motor. Dispone de 80 horas por semana para producirlos y en los procesos se requieren: 4 h para A y solo 2 h para B. Sin embargo, debido a lo peligroso del material usado, son necesarias 5 h de preparación y limpieza para el motopr B y 2 h para A. El tiempo de preparación total disponible es 120 h por semana. Determinar el número de cada clase de motores a producir que dé como resultado la máxima ganancia.   

### Solución
Se plantea el enunciado de arriba en un programa lineal:

#### Variables de decisión
Cantidad de motores A ($x_1$) y B ($x_2$)

#### Función objetivo
Máximizar $3x_1 + 4x_2$

#### Restricciones
Sujeto a (considerando la producción semanal):

Producción ($P1$): $4x_1 + 2x_2 \leq 80$ 

Preparación y limpieza ($P2$): $2x_1 + 5x_2 \leq 120$


```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

Graficamos la primera restricción ($P1$):

```python
plt.figure()
    
# Restricción para P1
plt.plot([0, 80.0/4.0], [80.0/2.0, 0], color='orange', alpha=0.25)
plt.fill_between([0, 80.0/4.0], [80.0/2.0, 0], color='orange', alpha=0.25, label='restricción de P1')

plt.xlabel('Motores A')
plt.ylabel('Motores B')
plt.legend()
plt.show()
```

Graficamos la segunda restricción ($P2$):

```python
plt.figure()
    
# Restricción para P1
plt.plot([0, 80.0/4.0], [80.0/2.0, 0], color='orange', alpha=0.25)
plt.fill_between([0, 80.0/4.0], [80.0/2.0, 0], color='orange', alpha=0.25, label='restricción de P1')

# Restricción para P2
plt.plot([0, 120.0/2.0], [120.0/5.0, 0], color='purple', alpha=0.25)
plt.fill_between([0, 120.0/2.0], [120.0/5.0, 0], color='purple', alpha=0.25, label='restricción de P2')

plt.xlabel('Motores A')
plt.ylabel('Motores B')
plt.legend()
plt.show()
```

Luego, la región factible de nuestro programa es la intersección de las restricciones $P1$ y $P2$:

Por ejemplo, el punto $(10,10)$ está dentro de la region factible

```python
plt.figure()
    
# Restricción para P1
plt.plot([0, 80.0/4.0], [80.0/2.0, 0], color='orange', alpha=0.25)
plt.fill_between([0, 80.0/4.0], [80.0/2.0, 0], color='orange', alpha=0.25, label='restricción de P1')

# Restricción para P2
plt.plot([0, 120.0/2.0], [120.0/5.0, 0], color='purple', alpha=0.25)
plt.fill_between([0, 120.0/2.0], [120.0/5.0, 0], color='purple', alpha=0.25, label='restricción de P2')

plt.xlabel('Motores A')
plt.ylabel('Motores B')

plt.scatter([10], [10])  # punto factible

plt.xlim(0)
plt.ylim(0)

plt.legend()
plt.show()
```

Graficamos la curva de nivel de algunos puntos que esten dentro de la región factible $(10,10) \ $ y $(10,\frac{70}{4})$

```python
plt.figure()
    
# Restricción para P1
plt.plot([0, 80.0/4.0], [80.0/2.0, 0], color='orange', alpha=0.25)
plt.fill_between([0, 80.0/4.0], [80.0/2.0, 0], color='orange', alpha=0.25, label='restricción de P1')

# Restricción para P2
plt.plot([0, 120.0/2.0], [120.0/5.0, 0], color='purple', alpha=0.25)
plt.fill_between([0, 120.0/2.0], [120.0/5.0, 0], color='purple', alpha=0.25, label='restricción de P2')

plt.xlabel('Motores A')
plt.ylabel('Motores B')

plt.scatter([10], [10])  # punto factible
plt.plot([0, 70.0/3], [70.0/4, 0], color='blue', linestyle='-.')
plt.scatter([10], [70/4])  # punto factible
plt.plot([0, 100.0/3], [100.0/4, 0], color='red', linestyle='-.')

plt.xlim(0)
plt.ylim(0)

plt.legend()
plt.show()
```

Notamos, de graficar distintas curvas de nivel, que el gradiente de la función va "hacia arriba", es decir el punto máximo será la intersección entre $4x_1 + 2x_2 = 80 \ $ y $ \ 2x_1 + 5x_2 = 120$,  es decir en el punto  $(10,20)$

```python
# Encontrando la intersección de las restricciones:
# Coeficientes de las ecuaciones de la forma a*x_1 + b*x_2 = c
# 4x_1 + 2x_2 = 80
# 2x_1 + 5x_2 = 120
a1, b1, c1 = 4, 2, 80  # Primera recta
a2, b2, c2 = 2, 5, 120  # Segunda recta

# Matriz de coeficientes
A = np.array([[a1, b1], [a2, b2]])

# Vector de constantes
B = np.array([c1, c2])

# Resolver el sistema de ecuaciones
solution = np.linalg.solve(A, B)

# Mostrar la solución (punto de intersección)
x_1, x_2 = solution
print(f"La intersección de las dos rectas es en el punto: $x_1$ = {x_1}, $x_2$ = {x_2}")
```
